# bar chart to compare different attributes of companies--using --xticks.

```python
profit=[10,25,35,15]
revenue=[100,200,150,50]
company=(['TCS','WIPRO','INFOYSYS','CAPGEMINI'])
pos=np.arange(len(company))
print(pos)
plt.xticks(pos,company)
plt.bar(pos-0.2,height=profit,width=0.5,label="Profit")
plt.bar(pos+0.2,height=revenue,width=0.5,label="Revenue")
plt.legend()
plt.xlabel("company")
plt.ylabel("profit")
plt.show()
```
